```python
from transformers import pipeline
```

```python
unmasker = pipeline('fill-mask', model='bert-base-uncased')
```

```python
text = "The movie was full of [MASK]."
result = unmasker(text)
print(result)
```

```python
for i, item in enumerate(result, start=1):
  print(f"{i}番目：{item['sequence']}")
```
