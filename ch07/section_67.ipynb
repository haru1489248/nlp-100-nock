## 精度の計測
精度の計測は混同行列から算出できる
混同行列の内容復習
- TP (True Positive)：正解=1 を 1 と予測（当たり）
- TN (True Negative)：正解=0 を 0 と予測（当たり）
- FP (False Positive)：正解=0 を 1 と予測（誤検知）
- FN (False Negative)：正解=1 を 0 と予測（見逃し）

精度の計測の種類
1. 正解率（Accuracy）
- なんで使う？
- 「全体としてどれくらい当たってるか」を一発で把握できる、分かりやすい指標。
- ただし クラス不均衡（例：99%が0）だと、何も考えず全部0予測でもAccuracyが99%になってしまうので注意。

$$
Accuracy = \frac{TP + TN}{TP+TN+FP+FN}
$$
2. 適合率（Precision）
- なんで使う？
- 「1だと予測したもののうち、どれだけ本当に1だったか」 を見たいときに使う。
- つまり 誤検知（FP）を減らしたい 場面で重要。
例：スパム判定で、普通のメールをスパム扱い（FP）すると困る → Precision重視。
$$
Precision = \frac{TP}{TP+FP}
$$
3. 再現率（Recall）
- なんで使う？
- 「本当は1だったものを、どれだけ取り逃さず拾えたか」 を見たいときに使う。
- つまり 見逃し（FN）を減らしたい 場面で重要。
- 例：病気検出で見逃し（FN）があると致命的 → Recall重視。
- ※ Recall は 感度 (Sensitivity) と呼ばれることもあります。
$$
Recall = \frac{TP}{TP+FN}
$$
4. F1スコア（F1-score）
- なんで使う？
- Precision と Recall は トレードオフ になりやすい（見逃しを減らそうとすると誤検- 知が増えたり、その逆も）。
- F1はそれを1つの値にまとめて、「バランスよく良いか」を見られる。
- 特に クラス不均衡 のとき、Accuracyより現実的な評価になりやすい。
- クラス不均衡とは正解ラベルが偏っていること
$$
F1 = \frac{2\cdot Precision\cdot Recall}{Precision + Recall} \\[6pt]
$$
$$
F1 = \frac{2TP}{2TP+FP+FN}
$$
