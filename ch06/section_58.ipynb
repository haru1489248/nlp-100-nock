# Ward法による階層化クラスタリングとデンドログラムでの表示

```python
import gensim.downloader as api
from scipy.cluster.hierarchy import dendrogram, ward
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# モデルをダウンロード（既にダウンロード済みならスキップされます）
print("ダウンロード中...")
vectors = api.load("word2vec-google-news-300")
print("完了！")
```

```python
# Ward法（問題58）
data_src = Path('/content/nlp-100-nock/ch06/assets/questions-words.txt')
target_section = ["capital-common-countries", "capital-world"]

data = []
current_section = ""
with open(data_src, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.split()
        if len(line) >= 2 and line[0] == ":":
            current_section = line[1]  # セクション名を取得
        elif current_section in target_section and len(line) == 4:
            for word in line:
                if word in vectors:
                    data.append(vectors[word])
```

```python
Z = ward(data)
```

```python
dendrogram(Z)
```
