```python
import gensim.downloader as api

# Google News Word2Vec モデルをダウンロード（数分かかります）
print("ダウンロード中...")
model = api.load("word2vec-google-news-300")
print("完了！")

# "United States" のベクトルを表示
vector = model["United_States"]
print(f"次元数: {len(vector)}")
print(f"ベクトル:\n{vector}")
```

```python
"""Ward法（ウォード法）"""

from scipy.cluster.hierarchy import dendrogram, ward
from pathlib import Path
from gensim.models import KeyedVectors
import matplotlib.pyplot as plt

model_src = Path('ch06/assets/GoogleNews-vectors-negative300.bin.gz')
data_src = Path('ch06/assets/questions-words.txt')

vectors = KeyedVectors.load_word2vec_format(model_src, binary=True)

target_section = ["capital-common-countries", "capital-world"]

data = []
for target in target_section:
    with open(data_src, 'r', encoding='utf-8') as f:
       for line in f:
           line = line.split()
           if line[1].startswith(target):
               current_section = line[1]
           elif line and current_section == target:
               if len(line) == 4:
                   for word in line:
                       data.append(vectors[word])

Z = ward(data)

dendrogram(Z)
plt.show()

# TODO: ユークリッド距離の勉強をする
```

```python
import gensim.downloader as api
from scipy.cluster.hierarchy import dendrogram, ward
from pathlib import Path
import matplotlib.pyplot as plt

# モデルをダウンロード
print("ダウンロード中...")
vectors = api.load("word2vec-google-news-300")
print("完了！")

# "United States" のベクトルを表示（問題60）
print(f"次元数: {len(vectors['United_States'])}")
print(f"ベクトル:\n{vectors['United_States']}")

# Ward法（問題58）
data_src = Path('/content/nlp-100-nock/ch06/assets/questions-words.txt')
target_section = ["capital-common-countries", "capital-world"]

data = []
current_section = ""
for target in target_section:
    with open(data_src, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.split()
            if len(line) > 0 and line[0].startswith(":"):
                current_section = line[0][1:]
            elif current_section == target and len(line) == 4:
                for word in line:
                    if word in vectors:
                        data.append(vectors[word])

Z = ward(data)
dendrogram(Z)
plt.show()
```
